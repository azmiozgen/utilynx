# Regression with real data


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
SEED = 73
np.random.seed(SEED)
```

## Get data

```python
csv_file = './data/insurance.csv'
df = pd.read_csv(csv_file)
print("# of rows: ", len(df))
df.head()
```

### Analyze data

```python
# Check NaN
df.isnull().sum()
```

```python
# See counts of sex, children, smoker, and region
print(df['sex'].value_counts())
print(df['children'].value_counts())
print(df['smoker'].value_counts())
print(df['region'].value_counts())
```

```python
# See histogram of age
df['age'].hist(color='blue')
plt.show()
```

```python
df['bmi'].hist(color='red')
plt.show()
```

```python
df['charges'].hist(color='orange')
plt.show()
```

```python
df_train = df.sample(frac=0.8, random_state=SEED)
df_test = df.drop(df_train.index)
print("# of train rows: ", len(df_train))
print("# of test rows: ", len(df_test))
```
